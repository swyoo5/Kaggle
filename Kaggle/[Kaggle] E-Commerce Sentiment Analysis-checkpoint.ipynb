```python
pattern = "http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\(\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+"
# 1. http[s]? : s ë’¤ì— ?ê°€ ìˆìœ¼ë¯€ë¡œ sëŠ” ì„ íƒì ì´ë‹¤.
# 2. :// : URLì—ì„œ í”„ë¡œí† ì½œê³¼ ë‚˜ë¨¸ì§€ ë¶€ë¶„ì„ êµ¬ë¶„í•˜ëŠ” ì—­í• ì„ í•œë‹¤.
# 3. (?: ...) : ì´ ê·¸ë£¹ ì•ˆì— ìˆëŠ” íŒ¨í„´ì€ ê·¸ë£¹í™”ë§Œ í•˜ê³  ìº¡ì³ëŠ” í•˜ì§€ ì•ŠëŠ”ë‹¤.
# (http|https) (?:http|https) ë¹„êµ
# ì „ìëŠ” http, https ë‘˜ì¤‘ í•˜ë‚˜ì— ë§¤ì¹˜ëœë‹¤.
# í›„ìëŠ” ì „ìì™€ ë§ˆì°¬ê°€ì§€ì´ì§€ë§Œ ì¶”ê°€ì ìœ¼ë¡œ ë§¤ì¹˜ëœ ë¶€ë¶„ì€ ê·¸ë£¹ìœ¼ë¡œ ê¸°ì–µë˜ì§€ ì•ŠëŠ”ë‹¤. ì¦‰, ì¶”í›„ì— ì´ ë¶€ë¶„ì„ ì–»ì–´ì˜¬ ìˆ˜ ì—†ë‹¤.

# 4. [a-zA-Z0-9$-_@.&+] : ì•ŒíŒŒë²³ ëŒ€ì†Œë¬¸ì, ìˆ«ì, íŠ¹ì • ê¸°í˜¸
# 5. [!*\(\),] : íŠ¹ìˆ˜ê¸°í˜¸ ì¤‘ ì¼ë¶€ê°€ ë§¤ì¹˜
# 6. (?:%[0-9a-fA-F][0-9a-fA-F]) : URL ì¸ì½”ë”© ë¬¸ìì—´ì„ ë‚˜íƒ€ë‚¸ë‹¤. "%20"ê³¼ ê°™ì€ í˜•ì‹ì˜ ì¸ì½”ë”©ëœ ë¬¸ìë¥¼ ì°¾ëŠ”ë‹¤.
# 7. + : ì•ì˜ íŒ¨í„´ì´ ìµœì†Œ í•œë²ˆ ì´ìƒ ë°˜ë³µë¨ì„ ì˜ë¯¸í•œë‹¤.
```

```python
def re_hiperlinks(text_list) :
    pattern = "http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\(\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+"
    return [re.sub(pattern, ' link ', r) for r in text_list]
```

```python
reviews_hiperlinks = re_hiperlinks(reviews_breakline)
df_comments["re_hiperlinks"] = reviews_hiperlinks
reviews_hiperlinks
```

# ë‚ ì§œ
ì˜¨ë¼ì¸ì—ì„œ êµ¬ë§¤í•œ ë¬¼ê±´ì˜ ì†Œë¹„ìì˜ ë¦¬ë·°ë¥¼ ë‹¤ë£¨ëŠ” í•œ ë‚ ì§œ ì–¸ê¸‰ì€ ë§¤ìš° í”í•˜ë‹¤. ì˜ˆì œë¥¼ ì‚´í´ë³´ê³  ì •ê·œí‘œí˜„ì‹ì„ ì ìš©í•´ ë‚ ì§œë¥¼ "data"ë¡œ ë°”ê¿”ë³´ì.

```python
def re_dates(text_list) :
    pattern = "([0-2][0-9]|(3)[0-1])(\/|\.)(((0)[0-9])|((1)[0-2]))(\/|\.)\d{2,4}"
    return [re.sub(pattern, " data ", r) for r in text_list]
```

```python
# í•˜ì´í¼ë§í¬ë¥¼ ì „ì²˜ë¦¬í•œ ë°ì´í„°ë“¤ì„ ì¶”ê°€ë¡œ ë‚ ì§œ ì „ì²˜ë¦¬
reviews_dates = re_dates(reviews_hiperlinks) 

# ë‚ ì§œ ì „ì²˜ë¦¬í•œ ì—´ì„ ë°ì´í„°í”„ë ˆì„ì— ì¶”ê°€
df_comments["re_dates"] = reviews_dates
```

# ì¶”ê°€ì ì¸ ê³µë°±ë¬¸ì
ì•ì„  ê³¼ì •ì„ ëª¨ë‘ ê±°ì¹œ í›„, ë¶ˆí•„ìš”í•œ ê³µë°±ë¬¸ìë¥¼ ì œê±°í•¨ìœ¼ë¡œì¨ í…ìŠ¤íŠ¸ë¥¼ ì •ì œí•˜ëŠ” ê²ƒì€ ë§¤ìš° ì¤‘ìš”í•˜ë‹¤.

```python
def re_whitespaces(text_list) :
    white_spaces = [re.sub('\s+', ' ', r) for r in text_list]
    white_spaces_end = [re.sub('[ \t]+$', '', r) for r in white_spaces]
    return white_spaces_end
```

```python
regex_transformers = {
    "break_line" : re_breakline,
    "hiperlinks" : re_hiperlinks,
    "dates" : re_dates,
    "money" : re_money,
    "numbers" : re_numbers,
    "negation" : re_negation,
    "special_chars" : re_special_chars,
    "whitespaces" : re_whitespaces
}

# max_features : TF-IDFë¡œ ë³€íšëœ í–‰ë ¬ì—ì„œ ì‚¬ìš©í•  ìµœëŒ€ íŠ¹ì§•ì˜ ìˆ˜ë¥¼ ì§€ì •í•œë‹¤.
# min_df : ë‹¨ì–´ê°€ ë¬¸ì„œì—ì„œ ë‚˜íƒ€ë‚˜ëŠ” ìµœì†Œ ë¹ˆë„ë¥¼ ë‚˜íƒ€ë‚¸ë‹¤. ì´ ê°’ë³´ë‹¤ ë‚®ì€ ë¹ˆë„ë¡œ ë‚˜íƒ€ë‚˜ëŠ” ë‹¨ì–´ëŠ” ë¬´ì‹œëœë‹¤. 
# ì—¬ê¸°ì—ì„œ, 7ê°œ ì´í•˜ì˜ ë¬¸ì„œì—ì„œ ë‚˜íƒ€ë‚˜ëŠ” ë‹¨ì–´ë“¤ì€ ë¬´ì‹œëœë‹¤.
# max_df : ë‹¨ì–´ê°€ ë¬¸ì„œì—ì„œ ë‚˜íƒ€ë‚˜ëŠ” ìµœëŒ€ ë¹ˆë„ë¥¼ ë‚˜íƒ€ë‚¸ë‹¤. ì´ ê°’ë³´ë‹¤ ë†’ì€ ë¹ˆë„ë¡œ ë‚˜íƒ€ë‚˜ëŠ” ë‹¨ì–´ëŠ” ë¬´ì‹œëœë‹¤.
# ì—¬ê¸°ì—ì„œ, 80% ì´ìƒì˜ ë¬¸ì„œì—ì„œ ë‚˜íƒ€ë‚˜ëŠ” ë‹¨ì–´ë“¤ì€ ë¬´ì‹œëœë‹¤.
# stop_words : ë¶ˆìš©ì–´ë¡œ ì²˜ë¦¬í•  ë‹¨ì–´ ëª©ë¡ì„ ì§€ì •í•œë‹¤.
# ì—¬ê¸°ì„œëŠ” pt_stopwords ê°€ ì‚¬ìš©ëœë‹¤.
vectorizer = TfidfVectorizer(max_features = 300, min_df = 7, max_df = 0.8, stop_words = pt_stopwords)

text_pipeline = Pipeline([
    ("regex", ApplyRegex(regex_transformers)),
    ("stopwords", StopWordsRemoval(stopwords.words("portuguese"))),
    ("stemming", StemmingProcess(RSLPStemmer())),
    ("text_features", TextFeatureExtraction(vectorizer))
])
```
