# 데이터 정제

```python
# HTML 태그 제거, 대괄호 제거, URL 제거, 볼용어 제거

# HTML 텍스트에서 HTML 태그를 제거하여 순수한 텍스트를 얻기 위한 함수
def strip_html(text) :
    # BeautifulSoup 객체 생성
    # BeautifulSoup : HTML 및 XML 문서를 파싱하고 검색하는 python 라이브러리
    soup = BeautifulSoup(text, "html.parser")
    # BeautifulSoup 객체에서 모든 태그를 제거하고 텍스트만 추출한다.
    # get_text() : HTML 태그가 제거된 순수한 텍스트를 반환한다.
    return soup.get_text()

# 대괄호 제거
def remove_between_square_brackets(text) :
    return re.sub("\[[^]]*\]", "", text)

# URL 제거
def remove_URL(text) :
    return re.sub(r'http\S+', '', text)

# 불용어 제거
def remove_stopwords(text) :
    final_text = []
    for i in text.split() :
        if i.strip().lower() not in stop : # 양쪽 공백 제거, 소문자화 
            final_text.append(i.strip())
    return " ".join(final_text)

# 노이즈 텍스트 제거(위의 함수 모두 적용)
def denoise_text(text) :
    text = strip_html(text)
    text = remove_between_square_brackets(text)
    text = remove_stopwords(text)
    return text

# text 열에 모두 적용
df["text"] = df["text"].apply(denoise_text)
df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
def get_top_text_ngrams(corpus, n, g) : # 상위 n개의 g개 단어 추출
    # n-gram의 범위를 (g, g)로 설정하여 해당 범위의 n-gram을 추출하도록 설정, 주어진 corpus에 대해 이 Vectorizer 학습시킨다.
    # ngram_range는 (min, max)로 주어지며 min에서 max까지의 n-gram을 생성하라는 의미이다.
    vec = CountVectorizer(ngram_range = (g, g)).fit(corpus)
    
    # 학습된 vectorizer를 사용하여 코퍼스의 문서를 벡터로 변환한다. 여기서는 문서에 포함된 n-gram의 빈도를 계산한다.
    bag_of_words = vec.transform(corpus)
    
    # 얻어진 벡터의 열별 합을 계산한다. 이는 각 n-gram의 전체 코퍼스에서의 빈도를 나타낸다.
    sum_words = bag_of_words.sum(axis = 0)
    
    # words_freq : 각 n-gram과 그 빈도를 저장한 리스트 
    # vec.vocabulary_.items() : CountVectorizer 객체에서 단어와 해당 단어의 인덱스를 포함하는 딕셔너리를 나타낸다. 단어-인덱스를 키-값으로 가지고 있다.
    # 예를 들어 apple이 전체 텍스트에서 5번째로 등장하면 vec.vocabulary_ 딕셔너리에는 {"apple" : 4}가 된다.
    # vec.vocabulary_.items()는 이 딕셔너리의 모든 키-값 쌍을 포함하는 리스트를 반환한다.
    words_freq = [(word, sum_words[0, idx]) for word, idx in vec.vocabulary_.items()]
    
    # 빈도에 따라 내림차순으로 정렬된 n-gram 리스트를 반환한다.
    words_freq = sorted(words_freq, key = lambda x : x[1], reverse = True)
    
    return words_freq[:n]
```

# 훈련 데이터, 테스트 데이터 분할

```python
# x에 text 데이터, y에 category 데이터 할당
x_train, x_test, y_train, y_test = train_test_split(df.text, df.category, random_state = 0)
```

```python
max_features = 1000
maxlen = 300
```

```python
# num_words : 토크나이저가 고려할 최대 단어의 수를 나타내는 매개변수
# 예를 들어, 수백만개의 단어가 있다면 모든 단어를 사용하는 것은 비효율적이다. 따라서 상위 몇천, 몇만개의 단어만 골라 모델을 학습시킨다.
tokenizer = text.Tokenizer(num_words = max_features) # 1000개의 단어를 골라 학습시킨다.
tokenizer.fit_on_texts(x_train) # x_train 데이터 학습
tokenized_train = tokenizer.texts_to_sequences(x_train) # x_train의 텍스트를 정수 시퀀스로 변환
x_train = sequence.pad_sequences(tokenized_train, maxlen = maxlen) # maxlen 길이에 맞춰 패딩
```

```python
tokenized_test = tokenizer.texts_to_sequences(x_test)
X_test = sequence.pad_sequences(tokenized_test, maxlen = maxlen)
```

```python
# GloVe 임베딩 파일 경로 지정
# 사전 학슴된 임베딩 벡터를 이용해 모델을 학습한다.
EMBEDDING_FILE = "C:/Users/User/OneDrive/바탕 화면/프로그래밍/자연어처리/Kaggle 필사/데이터/glove.twitter.27B.100d.txt"
```

```python
# GloVe 임베딩 파일에서 한 줄씩 읽어서 단어와 해당하는 임베딩 벡터를 추출
# def get_ceefs("apple", 0.1, 0.2, 0.3) => word = "apple", arr = (0.1, 0.2, 0.3)
# word : 단어, arr : 임베딩 벡터
def get_coefs(word, *arr) :
    return word, np.asarray(arr, dtype = "float32")
# GloVe 임베딩 파일에서 단어와 임베딩벡터를 추출하여 저장
# 단어 - 임베딩벡터 조함으로 이루어진 파일 => 단어 : 임베딩 벡터 딕셔너리로 만듬
embeddings_index = dict(get_coefs(*o.rstrip().rsplit(' ')) for o in open(EMBEDDING_FILE, "rt", encoding = "UTF8"))
```

```python
# 모든 임베딩 벡터를 행렬로 stack
# values : 임베딩 벡터
# embedding_index 딕셔너리의 임베딩벡터만 가져와서 np.ndarray 형식으로 저장
all_embs = np.stack(list(embeddings_index.values()))
# 전체 임베딩 벡터의 평균과 표준편차
emb_mean, emb_std = all_embs.mean(), all_embs.std()
# 임베딩 벡터의 크기
embed_size = all_embs.shape[1]
# 토크나이저를 사용하여 텍스트에서 단어를 숫자로 매핑
word_index = tokenizer.word_index
# 사용할 최대 단어 수
nb_words = min(max_features, len(word_index))
# 최종적으로 사용할 임베딩 행렬로 초기값은 평균과 표준편차를 사용하여
# 정규분포에서 랜덤하게 생성된다. 그 후 GloVe에서 가져온 임베딩 벡터로
# 대체된다. 최대 단어수에 도달하거나 해당 단어가 GloVe에 없는 경우 초기값 유지
embedding_matrix = np.random.normal(emb_mean, emb_std, (nb_words, embed_size))
for word, i in word_index.items() : # 단어, 단어의 인덱스
    if i >= max_features : continue # 고려할 단어의 최대 갯수에서 벗어나면
    embedding_vector = embeddings_index.get(word)
    if embedding_vector is not None : embedding_matrix[i] = embedding_vector
```

```python
batch_size = 256
epochs = 10
embed_size = 100
```

```python
# 케라스에서 제공하는 콜백 중 하나로 학습중에 학습률을 동적으로 조정하기 위해 사용된다.
# 지정된 모니터 인자에 대해 성능이 향상되지 않을 때 학습률을 조정하는 기능을 수행한다.
# monitor : 모니터링할 지표 "val_loss", "val_accuracy" 등
# factor : 학습률을 감소시키는 비율. 측정 지표가 개선되지 않으면 현재 학습률에 이 값을 곱하여 갱신한다.
# patience : 성능 향상이 없을 때 학습률을 감소시키기 전에 기다리는 에폭 수. patience 번의 에폭만큼 성능 향상이 없으면 학습률을 감소시킨다.
# min_lr : 학습률의 하한을 정한다. 학습률이 이 값보다 작아지지 않도록 한다.
learning_rate_reduction = ReduceLROnPlateau(monitor = "val_accuracy", patience = 2, verbose = 1, factor = 0.5, min_lr = 0.00001)
```

# 모델 훈련

```python
import tensorflow as tf
model = Sequential()
model.add(Embedding(max_features, output_dim = embed_size, weights = [embedding_matrix], input_length = maxlen, trainable = False))
model.add(LSTM(units = 128, return_sequences = True, recurrent_dropout = 0.25, dropout = 0.25))
model.add(LSTM(units = 64, recurrent_dropout = 0.1, dropout = 0.1))
model.add(Dense(units = 32, activation = "relu"))
model.add(Dense(1, activation = "sigmoid"))
model.compile(optimizer = tf.keras.optimizers.Adam(learning_rate = 0.01), loss = "binary_crossentropy", metrics = ["accuracy"])
```

```python
model.summary()
```

```python
history = model.fit(x_train, y_train, batch_size = batch_size, validation_data = (X_test, y_test), epochs = epochs, callbacks = [learning_rate_reduction])
```

# 모델 훈련 후 분석

```python
print("Accuracy of the model on Training Data is - ", model.evaluate(x_train, y_train)[1] * 100, "%")
print("Accuracy of the model on Testing Data is - ", model.evaluate(X_test, y_test)[1] * 100, "%")
```

```python
epochs = [i for i in range(10)]
fig, ax = plt.subplots(1, 2)
train_acc = history.history["accuracy"]
train_loss = history.history["loss"]
val_acc = history.history["val_accuracy"]
val_loss = history.history["val_loss"]
fig.set_size_inches(20, 10)

ax[0].plot(epochs, train_acc, "go-", label = "Training Accuracy")
ax[0].plot(epochs, val_acc, "ro-", label = "Testing Accuracy")
ax[0].set_title("Training & Testing Accuracy")
ax[0].legend()
ax[0].set_xlabel("Epochs")
ax[0].set_ylabel("Accuracy")

ax[1].plot(epochs, train_loss, "go-", label = "Training Loss")
ax[1].plot(epochs, val_loss, "ro-", label = "Testing Loss")
ax[1].set_title("Training & Testing Loss")
ax[1].legend()
ax[1].set_xlabel("Epochs")
ax[1].set_ylabel("Loss")
plt.show()
```

```python
pred = model.predict(X_test)
pred_binary = (pred > 0.5).astype(int)
pred_binary[:5]
```

```python
# classification_report : scikit-learn 라이브러리에서 제공되는 분류 모델의 성능을 평가하기 위한 함수
# y_test : 실제 레이블 값이 있는 테스트 데이터의 레이블 벡터
# pred_binary : 분류 모델의 예측 결과로 실제 레이블과 비교하여 평가된다.
# target_names : 클래스 레이블의 이름을 나타내는 리스트이다.
print(classification_report(y_test, pred_binary, target_names = ["Fake", "Not Fake"]))
```
