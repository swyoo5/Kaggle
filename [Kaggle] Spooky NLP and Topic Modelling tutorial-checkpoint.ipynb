### BoW 접근법
이 접근법은 단어의 개수를 시작 블륵으로 사용하고 해당 특정 단어와 관련된 벡터에 전체 텍스트에서 각 단어의 발생을 기록한다. 예를 들어 "I love to eat Burgers", "I love to eat Fries"와 같은 두 문장이 주어졌을 때 먼저 토큰화 해서 얻을 수 있는 6개의 단어를 얻는다. \[I, love, to, eat, Burgers, Fries\]

BoW 접근법을 통해 텍스트를 벡터화 하면 6개의 벡터를 얻는다.

```python
sentence = ["I love to eat Burgers",
            "I love to eat Fries"]
vectorizer = CountVectorizer(min_df = 0)
sentence_transform = vectorizer.fit_transform(sentence)
```

```python
print("The features are : \n {}".format(vectorizer.get_feature_names()))
print("\nThe vectorized array looks like : \n {}".format(sentence_transform.toarray()))
```

벅테화된 텍스트의 출력으로부터 우리는 feature들이 vectorizer에 입력한 텍스트의 말뭉치 안의 단어로 구성돼있는 것을 알 수 있다.

변형된 텍스트에 대하여 간단히 호출하여 값을 검사하고 싶을 수 있다. 하지만 호출을 하려고 했을 때 "sparse matrix of type class 'numpy.int64' with 8 stored elements in Compressed Sparse Row format"(압축된 희소 행 형식으로 8개의 저장된 원소가 있는 numpy.int64 유형 클래스의 희소 행렬)이러한 메시지가 뜬다. 이 의미는 vectorizer가 변형된 텍스트를 대부분의 값이 0이거나 무시해도 될만한 값인 행렬로 반환한다는 뜻이다.

# 3. 토픽 모델링
1. Latent Drichlet Allocation : 코퍼스의 단어에 가중치를 할당하여 데이터에 잠재된 주제를 찾아내는 확률적 생성형 모델. 여기서 각 주제는 각 단어에 서로 다른 확률 가중치를 할당한다.
2. Non-negative Matrix Factorization : 입력 행렬을 가져와서 이 행렬의 factorization을 두개의 다른 행렬로 근사하는 근사적인 방법. 단 행렬의 값은 음수가 아니라는 주의가 필요하다.

```python
# model : topic modeling에서 사용된 모델 객체(Latent Drichlet Allocation, Non Negative Matrix Factorization)
# feature_names : 각 특성의 이름이 담긴 리스트 또는 배열
# n_top_words : 각 주제별로 출력할 상위 단어의 개수
def print_top_words(model, feature_names, n_top_words) :
    
    # model.components_ : 모델에서 추정된 주제의 단어 분포를 나타내는 속성
    for index, topic in enumerate(model.components_) :
        # Topic 번호
        message = "\nTopic #{} : ".format(index)
        # topic.argsort()[...] : 상위 n_top_words 개수만큼의 단어의 인덱스를 내림차순으로 정렬한다.
        # 인덱스 -1, -2, ..., -n_top_words
        message += " ".join([feature_names[i] for i in topic.argsort()[:-n_top_words - 1 : -1]])
        print(message)
        print('=' * 80)
```

## 3a. 모든 전처리 과정을 합친다.
이전 절에서 했던 토큰화, 불용어 제거, stemming, lemmatizing 등등 과정을 다시 해야 하는가? 다행히도 토큰화, 불용어 제거 두가지 단계는 상위 레벨 구성요소에 포함되어있다. Sklearn에서 불용어 제거는 stop_words 인자에 "english"만 입력해주면 자동적으로 영어 불용어 리스트를 만들어낸다.

### CountVectorizer 클래스를 lemmatizer로 확장

```python
# WordNetLemmatizer 객체 생성
lemm = WordNetLemmatizer()
class LemmaCountVectorizer(CountVectorizer) : # 상속
    # CountVectorizer 클래스의 메서드 오버라이딩
    # 문서를 단어로 분리하는 함수
    def build_analyzer(self) :
        # 부모 클래스인 CountVectorizer의 build_analyzer 메서드를 호출하여 원래의 분석기를 가져온다.
        # super : 현재 클래스의 부모 클래스를 불러오는 함수이다.
        # LemmaCountVectorizer : 부모 클래스를 찾고자 하는 서브클래스 인자
        # self : 서브 클래스의 instance
        # build_analyzer() : 부모 클래스의 build_analyzer() 메서드를 의미한다.
        analyzer = super(LemmaCountVectorizer, self).build_analyzer()
        # 가져온 분석기를 이용하여 새로운 분석기를 정의한다.
        # 주어진 문서의 각 단어에 대해 레마타이제이션을 수행한다.
        # doc : 문서
        # analyzer(doc) : 입력 문서에 대해 CountVectorizer 클래스에서 상속된 원래 analyzer 메서드를 호출한다.
        # analyzer 메서드는 문서를 단어로 토큰화하는 역할을 한다.
        return lambda doc : (lemm.lemmatize(w) for w in analyzer(doc))
```

```python
text = list(train.text.values)
# max_df : 단어장에 포함되기 위한 최대 문서 빈도, 주어진 값 이상의 문서에서 발생한 단어는 무시된다.
# min_df : 주어진 값 미만의 문서에서 발생한 단어는 무시된다.
tf_vectorizer = LemmaCountVectorizer(max_df = 0.95, # 95% 이상의 문서에서 나타나는 단어 무시
                                    min_df = 2, # 2개 미만의 문서에서 발생한 단어 무시
                                    stop_words = "english",
                                    decode_error = "ignore")
tf = tf_vectorizer.fit_transform(text)
```

```python
# TF-IDF vectorizer에서 추출된 단어들의 이름을 가져온다.
feature_names = tf_vectorizer.get_feature_names()
# TF-IDF 벡터화된 문서들의 단어 빈도수를 합산
# axis = 0은 열방향으로 합산
# ravel() : 다차원 베열을 1차원 배열로 평탄화
count_vec = np.asarray(tf.sum(axis = 0)).ravel()
# 특징 이름과 해당 특징의 빈도수를 튜플로 묶어 리스트로 만든다.
zipped = list(zip(feature_names, count_vec))
# 빈도수에 따라 내림차순으로 정렬된 튜플 리스트를 빈도수와 특징 이름으로 분리한다.
# "*" 는 x, y에 따로 할당을 하도록 unpacking 해준다.
x, y = (list(x) for x in zip(*sorted(zipped, key = lambda x : x[1], reverse = True)))
Y = np.concatenate([y[0 : 15], y[-16 : -1]])
X = np.concatenate([x[0 : 15], x[-16 : -1]])

data = [go.Bar(
            x = x[0 : 50],
            y = y[0 : 50],
            marker = dict(colorscale = "Jet",
                         color = y[0 : 50]),
            text = "Word counts")]

fig = go.Layout(
    title = "Top 50 Word Frequencies after Preprocessing")

fig = go.Figure(data = data, layout = layout)
py.iplot(fig, filename = "basic-bar")
```

### 3b. Latent Drichlet Allocation


```python
# n_components : 문서 집합에서 몇개의 토픽을 찾을지 결정
# max_iter : 최적화를 위한 반복 횟수
# learning_method : 토픽 모델을 학습하는 방법 지정. "batch", "online"
# "batch" : 전체 데이터 세트를 사용하여 한번에 모델을 업데이트
# "online" : 일부 데이터를 사용하여 반복적으로 모델을 업데이트하며 대규모 데이터에 유용하다.
# learning_offset : "online" 학습 방법에서 학습 속도를 조절하는 파라미터이다. 높은 값은 초기 학습을 더 빠르게 하도록 한다.
# random_state : 시드값
lda = LatentDirichletAllocation(n_components = 11, max_iter = 5,
                                learning_method = "online",
                                learning_offset = 50.,
                                random_state = 0)
```

```python
lda.fit(tf)
```

```python
n_top_words = 40
print("\nTopics in LDA model : ")
tf_feature_names = tf_vectorizer.get_feature_names()
# model : topic modeling에서 사용된 모델 객체(Latent Drichlet Allocation, Non Negative Matrix Factorization)
# feature_names : 각 특성의 이름이 담긴 리스트 또는 배열
# n_top_words : 각 주제별로 출력할 상위 단어의 개수
print_top_words(lda, tf_feature_names, n_top_words) 
```

```python
first_topic = lda.components_[0]
second_topic = lda.components_[1]
third_topic = lda.components_[2]
fourth_topic = lda.components_[3]
print(first_topic)
print(tf_feature_names)
```

```python
# 각 주제에서 상위 50개 단어 추출
# topic.argsort()[:-50-1:-1] : 확률이 높은 50개 단어의 인덱스
first_topic_words = [tf_feature_names[i] for i in first_topic.argsort()[:-50-1:-1]]
second_topic_words = [tf_feature_names[i] for i in second_topic.argsort()[:-50-1:-1]]
third_topic_words = [tf_feature_names[i] for i in third_topic.argsort()[:-50-1:-1]]
fourth_topic_words = [tf_feature_names[i] for i in fourth_topic.argsort()[:-50-1:-1]]
print(first_topic_words)
```
